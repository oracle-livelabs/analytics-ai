## **Write the transformed and enhanced table to multiple databases for Anaytics or integration back to FDI** ##

```python
silver_supplier_feedback_AI_join_csv = "silver_supplier_feedback_AI_Join_csv"
```

```python
# Read managed silver tables
# Using spark.table
constructors_df = spark.table(f"silver_supplier.supplier_schema.silver_supplier_feedback_ai_join_csv")
```

```python
# optionally change the table column names

constructors_final_df = constructors_df \
    .withColumnRenamed("supplier_name_silver", "supplier_name")\
    .withColumnRenamed("supplier_continent_silver", "continent")\
    .withColumnRenamed("supplier_country_silver", "country")\
    .withColumnRenamed("supplier_city_silver", "city")\
    .withColumnRenamed("summary_silver", "summary_review")\
    .withColumnRenamed("Rating", "rating")
```

```python
# check the datafrome with new column names
try:
    constructors_final_df.show()

except ValueError as e:
    print("Invalid input:", e)
except ZeroDivisionError as e:
    print("Cannot divide by zero:", e)
except Exception as e: # Catch any other unexpected errors
    print("An unexpected error occurred:", e)
    print(type(e))  
```

```python
#  Create database schema and table in the supplier_external_26ai catalog to be inserted into

spark.sql("create schema if not exists supplier_external_26ai.admin")

spark.sql("create table if not exists supplier_external_26ai.admin.gold_supplier_feedback ( supplier_name VARCHAR(100), continent VARCHAR(100), country VARCHAR(100), city VARCHAR(100), summary_review VARCHAR(1000), rating VARCHAR(10) )")
```

```python
#  write the data frame into the database table
#  Using DataFrame overwrite
#  we could have used spark.sql,  sql, or data frame write or merge

try: 

   constructors_final_df.write.mode("overwrite").insertInto(f"supplier_external_26ai.admin.gold_supplier_feedback")
except ValueError as e:
    print("Invalid input:", e)
except ZeroDivisionError as e:
    print("Cannot divide by zero:", e)
except Exception as e: # Catch any other unexpected errors
    print("An unexpected error occurred:", e)
    print(type(e)) 
    
df = spark.read.table("supplier_external_26ai.admin.gold_supplier_feedback")
display (df)
```
