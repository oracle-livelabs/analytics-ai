```python
# Define parameters (can set parameters in a workflow job)
target_type   =oidlUtils.parameters.getParameter("TARGET_TYPE", "table")
target_format =oidlUtils.parameters.getParameter("TARGET_FORMAT", "adw")
silver_catalog    = "f1_demo_silver"
gold_catalog    = "f1_demo_gold"
adw_catalog = "f1_demo_adw"
silver_schema     = "f1_history"
gold_schema     = "f1_history"
adw_schema =      "demo_f1_history"
gold_table_dlt = "f1_constructors_dlt"
gold_table_par = "f1_constructors_par"
```

```python
# Step 1: Read managed silver tables
constructors_df = spark.table(f"{silver_catalog}.{silver_schema}.f1_constructors_dlt")
```

```python
# Step 2: Rename Columns
constructors_final_df = constructors_df \
    .withColumnRenamed("name", "team_name")\
    .withColumnRenamed("ingestion_date", "update_ts")
```

```python
constructors_final_df.show()
```

```python
if target_type == 'file':
    if target_format == 'parquet':
        constructors_final_df.write.mode("overwrite").parquet(f"{gold_folder_path}/constructors")
elif target_type == 'table':
    if target_format == 'parquet':
        constructors_final_df.write.mode("overwrite").format("parquet").saveAsTable(f"{gold_catalog}.{gold_schema}.f1_constructors_dlt")
    elif  target_format == 'delta':
        constructors_final_df.write.mode("overwrite").format("delta").saveAsTable(f"{gold_catalog}.{gold_schema}.f1_constructors_dlt")
    elif target_format == 'adw':
        constructors_final_df.write.insertInto(f"{adw_catalog}.{adw_schema}.f1_constructors")
```
