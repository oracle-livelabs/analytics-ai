```python
## This notebook ingest the raw "lap times" oracle redbull file from the bronze volume to bronze table
### Ingest lap_times.csv file
```

```python
# Define parameters (can set parameters in a workflow job)
target_type   =oidlUtils.parameters.getParameter("TARGET_TYPE", "table")
target_format =oidlUtils.parameters.getParameter("TARGET_FORMAT", "delta")
bronze_folder_path    = "/Volumes/f1_bronze/bronze/f1_bronze_volume"
bronze_catalog    = "f1_bronze"
bronze_schema     = "bronze"
bronze_table_dlt = "f1_lap_times_dlt"
bronze_table_par = "f1_lap_times_par"
```

# ----------
##### Step 1 - Read the CSV file using the spark dataframe reader
#  ----------

```python
from pyspark.sql.types import StructType, StructField, IntegerType, StringType

lap_times_schema = StructType([
    StructField("RACEID", IntegerType(), True),
    StructField("DRIVERID", IntegerType(), True),
    StructField("LAP", IntegerType(), True),
    StructField("POSITION", IntegerType(), True),
     StructField("TIME", StringType(), True),
    StructField("MILLISECONDS", IntegerType(), True)
])
```

```python
laptimes_df = spark.read \
.option("header", True) \
.schema(lap_times_schema) \
.csv(f"{bronze_folder_path}/lap_times.csv")
```

# ----------
##### Step 2 - Write the output to bronze catalog as parquet file or delta/parquet table
#  ----------

```python
if target_type == 'file':
    if target_format == 'parquet':
        laptimes_df.write.mode("overwrite").parquet(f"{bronze_folder_path}/lap_times_par")
elif target_type == 'table':
    if target_format == 'parquet':
        laptimes_df.write.mode("overwrite").format("parquet").saveAsTable(f"{bronze_catalog}.{bronze_schema}.{bronze_table_par}")
    elif  target_format == 'delta':
        laptimes_df.write.mode("overwrite").format("delta").saveAsTable(f"{bronze_catalog}.{bronze_schema}.{bronze_table_dlt}")
```
